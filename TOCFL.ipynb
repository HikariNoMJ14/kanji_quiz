```python
import pandas as pd
```

```python
def replace(x):
    if x == 'Novice 1' or x == 'Novice 2':
        y = 1.0
    elif x == 'A Level 1':
        y = 2.0
    elif x == 'A Level 2':
        y = 3.0
    elif x == 'B Level 1':
        y = 4.0
    elif x == 'B Level 2':
        y = 5.0
    elif x == 'C Level 1':
        y = 6.0
    else:
        print('error')

    return y
```

```python
tocfl = pd.read_csv('./data/TOCFL_en.csv', index_col=0)
tocfl['level 2'] = tocfl['level 2'].apply(replace)
tocfl = tocfl.drop(['level'], axis=1)
tocfl = tocfl[tocfl['level 2'] <= 4]
dictionary = pd.read_csv('./data/cedict_hsk_tocfl.txt',
    names=[
        'key', 'traditional', 'simplified',
        'pinyin', 'english', 'hsk',
        'hskCategory', 'tocfl', 'tocflCategory'
    ],
    delimiter='\t'
)[[
    'traditional',
    'english',
    'pinyin', 
    'tocfl',
    'tocflCategory'
]]
dictionary = dictionary[dictionary['tocflCategory'] == 'original']
dictionary = dictionary.drop('tocflCategory', axis=1)
tocfl_6 = pd.read_csv('./data/tocfl_6.txt', names=[
        'char', 'alt_char', 'level'
    ], delimiter='\t')
```

```python
tocfl = tocfl.reset_index()
tocfl['index'] = tocfl['index'] + 1
tocfl['char 2'] = tocfl['character'].str.replace(r'\(.*\)', '').str.replace('（.*）', '').str.strip()
tocfl['char 3'] = tocfl['char 2'].str.replace(r'\/.*', '').str.strip()
tocfl['zhuyin'] = tocfl['zhuyin'].str.replace('', '')
```

```python
a = pd.merge(
    left=tocfl, 
    right=dictionary, 
    how='left',
    left_on='char 3', 
    right_on='traditional'
)
```

```python
import re 

final_dict = {}
for i in range(1, a['number'].max() + 1):
    t = a[a['number'] == i]   
    if len(t) != 0:
        y = {
            "character": t.iloc[0,1],
            "zhuyin": t.iloc[0,2],
            "pinyin_1": t.iloc[0,3],
            "pinyin_2": t.iloc[0,4],
            "topic": t.iloc[0,6],
            "level": t.iloc[0,7],
            "part of speech": t.iloc[0,8]
        }

        eng = []
        for i, x in t.iterrows():
            w = x['english'].split('/')
            for j in range(min(len(w), 4)):
                z = w[j]
                print(t.iloc[0,1], z)
                z = re.sub('(CL\:[\u4e00-\u9fa5]\|*[\u4e00-\u9fa5]*)', '', z)
                z = re.sub(r'\[.*\]', '', z)
                print(t.iloc[0,1], z)
                print(z.strip(), '-------')
                
                if z.strip() != '' and z not in eng:
                    eng.append(z)

        y['english'] = list(eng)

        final_dict[t.iloc[0,0]] = y
    else:
        print(i)
            

final = pd.DataFrame(final_dict).T
```
